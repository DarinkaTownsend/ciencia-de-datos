```python
import pandas as pd
import numpy as pn
```

```python
archivo_entrada = "datos2022.csv"
archivo_salida = "datos2022_coma.csv"

# Leer el archivo de entrada
with open(archivo_entrada, 'r') as f:
    contenido = f.read()

# Reemplazar punto y coma por coma
contenido_modificado = contenido.replace(';', ',')

# Guardar el contenido modificado en el archivo de salida
with open(archivo_salida, 'w') as f:
    f.write(contenido_modificado)

print(f"Se ha procesado el archivo. Puedes encontrar el resultado en {archivo_salida}")
```

# Transformacion de datos



## Tipo movimiento

Entradas:1
Salidas: 2

```python
datos['tip_movi'] = datos['tip_movi'].replace({1: 'Entrada', 2: 'Salida'})
```

```python
datos
```

## Tipo nacimiento

1 = Ecuatorianos
2 = Extranjeros

```python
datos['tip_naci'] = datos['tip_naci'].replace({1: 'Ecuatorianos', 2: 'Extranjeros'})
```

```python
datos
```

## Sexo

Hombre:1
Mujer: 2
Indefinido: 3


```python
datos['sex_migr'] = datos['sex_migr'].replace({1: 'Hombre', 2: 'Mujer',3: "Indefinido"})
```
