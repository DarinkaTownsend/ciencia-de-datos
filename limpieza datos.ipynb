## Eliminar outliders

```python
cantidad_filas = len(datos)
print(f"Cantidad de filas antes: {cantidad_filas}")
```

```python
Q1 = datos['edad'].quantile(0.25)
Q3 = datos['edad'].quantile(0.75)

# Calcular el rango intercuartílico (IQR)
IQR = Q3 - Q1

# Definir los umbrales para outliers
umbral_inferior = Q1 - 1.5 * IQR
umbral_superior = Q3 + 1.5 * IQR

# Filtrar el DataFrame para eliminar outliers
datos = datos[(datos['edad'] > umbral_inferior) & (datos['edad'] < umbral_superior)]
```

```python
cantidad_filas = len(datos)
print(f"Cantidad de filas despues: {cantidad_filas}")
```

## Precios por viaje en continentes

```python

def asignar_precio(row):
    precios = {
        ('América', 'América'): 800.90,
        ('América', 'Europa'): 1200.98,
        ('América', 'Oceanía'): 1500.98,
        ('América', 'África'): 1400,
        ('América', 'Asia'): 1600,
        ('Europa', 'América'): 1100,
        ('Europa', 'Europa'): 1000,
        ('Europa', 'Oceanía'): 1300.70,
        ('Europa', 'África'): 1200,
        ('Europa', 'Asia'): 1400,
        ('Oceanía', 'América'): 1700,
        ('Oceanía', 'Europa'): 1400,
        ('Oceanía', 'Oceanía'): 1200,
        ('Oceanía', 'África'): 1600,
        ('Oceanía', 'Asia'): 1800.90,
        ('África', 'América'): 1600,
        ('África', 'Europa'): 1300,
        ('África', 'Oceanía'): 1800,
        ('África', 'África'): 1100,
        ('África', 'Asia'): 1700,
        ('Asia', 'América'): 1800,
        ('Asia', 'Europa'): 1500,
        ('Asia', 'Oceanía'): 1900,
        ('Asia', 'África'): 1800,
        ('Asia', 'Asia'): 1000,
    }
    # Devolver el precio correspondiente a la combinación de continentes
    return precios.get((row['cont_res'], row['cont_prod']), None)
```

```python
# Crear la nueva columna "precio_del_pasaje"
datos['precio_pasaje'] = datos.apply(asignar_precio, axis=1)
```
